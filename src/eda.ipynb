Now, in order to understand each column, I will construct an encoding strategy 

- Track Name
    - Going to use sent2vec because bagofwords term frequency will be too sparse - for a dataset as small as this one - to get meaningful results
- Artist Name(s)
    - I will map each unique artist to a unique number
- Album Name
    - Similar to 'Track Name' I will use sent2vec for the same reason
- Album Artist Name(s)
    - Can have up to seven artists on a particular tracks album, I will adopt the same strategy as Artist Name(s) and then use a one hot encoding to create seven columns. This means albums with only one artist on will look like [unique artists number here,-1,-1,-1,-1,-1,-1]
- Album Release Date
    - Take the earliest date as 0 and then use a datetime.timedelta to get the difference form this date in seconds. I may also add another column which is the month of the year, in case this has some relevance.
- Artist Genres
    - Determine all possible genres in the samples, and then use a onehot encoding to represent the variable (too much information under one variable)
- Label
    - I will map each unique label to a unique number

### Distributions
Below are some useful functions that allow faster analysis 

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

def pull_out_items(df, column, separator=','):
    names=[]
    for i in df[column].values:
        i = str(i).strip()
        if not separator in i:
            names.append(i)
        else :
            nme=i.split(separator)
            names.extend(nme)
    return names

def draw_counts_bar_chart(labels, counts, n, x_title='artist'):
    fig,ax=plt.subplots()
    ax.bar(labels[:n],counts[:n])
    ax.set_xticks(np.arange(0,n))
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
    ax.set_ylabel('Appearences on List')
    ax.set_xlabel(x_title)
    ax.grid()
    fig.tight_layout()
```

```python
#writing a function that allowed for the conversion from string to python date_time
#there are some tricky edge cases that need to be handled i.e nan, missing months/days
def fix_times(df,date_column):
    dates = []
    for j,i in enumerate(df[date_column]): 
        print(j)
        if str(i) == 'nan-01-01':
            dates.append('2000-01-01')
        elif str(i) == '0000-01-01':
            dates.append('2000-01-01')
        elif i == '0000':
            dates.append('2000-01-01')
        elif len(str(i)) == 4:
            dates.append(f'{i}-01-01')
        elif len(str(i)) == 7:
            dates.append(f'{i}-01')
        else:
            dates.append(i)
        
    df[date_column] = dates
    return df


#used pd.date_time to convert to date_time
df_dropped = fix_times(df_dropped,'Album Release Date')
df_dropped['Album Release Date'] = pd.to_datetime(df_dropped['Album Release Date'])
```

```python
#getting the min and max years to ascertain the decades in the dataset
min_year=df_dropped['Album Release Date'].min()
print (min_year)

max_year=df_dropped['Album Release Date'].max()
print (max_year)
```

```python
#function to get decade from the year
def get_decade(year):
    year=year.to_pydatetime()
    decade=str(year.year)[:3]+'0'
    return decade
```

```python
###### import matplotlib.pyplot as plt
import numpy as np

import matplotlib.patheffects as path_effects
import numpy as np

def create_annotated_donut_chart(names, counts, figsize=(8, 8), title="Decades"):
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=figsize)

    # Define a color palette
    colors = plt.cm.Spectral(np.linspace(0, 1, len(names))[::-1])

    # Plot the pie chart
    wedges, texts, autotexts = ax.pie(
        counts,
        labels=names,
        autopct=lambda p: '{:.1f}%'.format(p),
        textprops=dict(color="w", fontsize=20, path_effects=[path_effects.withStroke(linewidth=3, foreground='black')]),
        colors=colors,
        wedgeprops=dict(width=0.60, edgecolor='black')
    )

    # Bold edges
    for wedge in wedges:
        wedge.set_edgecolor('white')

    # Add a circle in the center to create a donut chart
    centre_circle = plt.Circle((0, 0), 0.2, color='white', fc='white', linewidth=0)
    ax.add_artist(centre_circle)

    # Set aspect ratio to be equal
    ax.axis('equal')


create_annotated_donut_chart(decade,counts)

```

```python
#function which filters the dataframe according to decade, and plots the top
#n most popular entries for that column
def plot_top_n_per_decade(df,column_name,decade,n=10,x_title=''):
    df_dec=df.loc[df['Decade']==decade]
    
    items = pull_out_items(df_dec, column_name, ',')
    cg=Counter(items)
    item,counts=zip(*cg.most_common())
    draw_counts_bar_chart(item, counts, n=n, x_title=x_title)
```

```python
import seaborn as sns

def ridgeplot(df,row='Decade',x='Tempo',title=''):

    sns.set_theme(style='white', rc={'axes.facecolor': (0, 0, 0, 0), 'axes.linewidth':2})
    palette = sns.color_palette('Set2', 12)
    g = sns.FacetGrid(df, row=row, hue=row, aspect=9, height=1.2,row_order=sorted(df[row].unique()))
    g.map_dataframe(sns.kdeplot, x=x, fill=True, alpha=1)
    g.map_dataframe(sns.kdeplot, x=x, color='black')
    
    def label(x, color, label):
        ax = plt.gca()
        ax.text(0, .2, label, color='black', fontsize=13,
                ha='left', va='center', transform=ax.transAxes)
    
    g.map(label,row)
    g.fig.subplots_adjust(hspace=-.5)
    g.set_titles('')
    g.set(yticks=[], xlabel=x, ylabel=None)
    g.despine( left=True)
    plt.suptitle(title, y=0.98)
```
